- **Stochasticity Check**

    Compute association matrix (i.e., number of times two nodes are assigned in the same community) across n_runs runs of Infomap Community Detection.

```python
def compute_asso(comm, mapping):
    n = len(mapping.items())
    A_asso = np.zeros((n, n))
    
    for k1, v1 in comm.items():
        for k2, v2 in comm.items():
            if v1 == v2:
                A_asso[mapping[k1], mapping[k2]] = 1
    
    return A_asso
```

```python
def robust_check(G, n_runs):
    
    A_asso = np.zeros((len(G.nodes), len(G.nodes)))
    mapping = {}
    
    for i in range(n_runs):
        G_comm = IP.process_infomap(G, resultPrint = False, convert = False, seed = i+1)
        if i == 0:
            r_G = IP.rearrange_graph(G, G_comm)
            mapping = {list(r_G.nodes)[i]:i for i in range(len(r_G.nodes))}
        A_asso += compute_asso(G_comm, mapping)
        #print(i)
    
    A_asso = np.true_divide(A_asso, n_runs)
    
    return A_asso, mapping
```

```python
A_asso1, mapping1 = robust_check(G1, 50)
```

```python
A_asso2, mapping2 = robust_check(G2, 50)
```

## 3.1. BT Component Size

```python
def get_bowtie_role(G, bowtie):
    S, IN, OUT, TUBES, INTENDRILS, OUTTENDRILS, OTHER, UNASSIGNED = [], [], [], [], [], [], [], []
    
    for node in G.nodes:
        role = G.nodes[node][bowtie]
        if role == 'S':
            S.append(node)
        elif role == 'IN':
            IN.append(node)
        elif role == 'OUT':
            OUT.append(node)
        elif role == 'TUBES':
            TUBES.append(node)
        elif role == 'INTENDRILS':
            INTENDRILS.append(node)
        elif role == 'OUTTENDRILS':
            OUTTENDRILS.append(node)
        elif role == 'OTHERS':
            OTHER.append(node)
        else:
            UNASSIGNED.append(node)
    
    if bowtie == 'within_role':
        return S, IN, OUT, TUBES, INTENDRILS, OUTTENDRILS, OTHER
    else:
        return S, IN, OUT, TUBES, INTENDRILS, OUTTENDRILS, OTHER, UNASSIGNED
```

```python
def merge_two_dicts(x, y):
    """Given two dictionaries, merge them into a new dict as a shallow copy."""
    z = x.copy()
    z.update(y)
    return z
```

## 3.2. BT Stat Sig

```python
import NetworkX_Extension as nx_e
import BowTie as bt
```

```python
def get_bowtie_components_rank(G):
    S, IN, OUT, TUBES, INTENDRILS, OUTTENDRILS, OTHER = bt.get_bowtie_components(G)
    bt_list = np.array([len(S), len(IN), len(OUT), len(TUBES), len(INTENDRILS), len(OUTTENDRILS), len(OTHER)])
    
    # Directed Configuration Model
    dcm_bt_list = []; dcm_bt_list.append(bt_list)
    G_indegree = [d for n, d in G.in_degree()]; G_outdegree = [d for n, d in G.out_degree()]
    for i in range(1000):
        G_dcm = nx.directed_configuration_model(G_indegree, G_outdegree, seed = i)
        #G_dcm = nx.DiGraph(G_dcm) # remove parallel edges
        G_dcm.remove_edges_from(nx.selfloop_edges(G_dcm)) # remove self-loops
        S, IN, OUT, TUBES, INTENDRILS, OUTTENDRILS, OTHER = bt.get_bowtie_components(G_dcm)
        dcm_bt_list.append(np.array([len(S), len(IN), len(OUT), len(TUBES), len(INTENDRILS), len(OUTTENDRILS), len(OTHER)]))
    dcm_bt_array = np.array(dcm_bt_list)
    
    # Calculate p-value
    rank = (np.argsort(dcm_bt_array, axis=0)[0])/1000 * 100
    
    return rank
```

```python
def bowtie_plot(percentage, rank, savename):
    fig, ax = plt.subplots(figsize=(15,10))
    plt.xlim(0, 1); plt.ylim(0, 1)

    size = percentage * 230 + 100
    color = rank
    cm = plt.cm.get_cmap("coolwarm"); 

    x = [0.5, 0.2, 0.8, 0.5, 0.2, 0.8, 0.5]; y = [0.5, 0.5, 0.5, 0.2, 0.8, 0.8, 0.8]
    plot = plt.scatter(x, y, s = size, c = color, cmap = cm, edgecolor = 'k', linewidth = 3)
    plot.set_clim(0,100)

    cbar = plt.colorbar(plot)
    cbar.ax.tick_params(labelsize=15)
    cbar.set_label('Rank (%)', rotation = -90, labelpad = 25, fontsize = 20)

    #cbar = plt.colorbar(plot, location="bottom")
    #cbar.ax.tick_params(labelsize=15)
    #cbar.set_label('Rank (%)', labelpad = 20, fontsize = 20)

    plt.axis('off');
    
    plt.savefig('Figures/' + savename, transparent = False)
```
